## 📋 Introduction  
In this project, I performed **data cleaning**, **feature engineering**, and **exploratory data analysis (EDA)** on the famous Titanic dataset. This notebook walks through advanced techniques for handling missing data, detecting outliers, creating new features, and visualizing survival patterns.  

- Dataset: Titanic Dataset from [Kaggle](https://www.kaggle.com/c/titanic/data)
- Tools: **Pandas**, **Seaborn**, **Matplotlib**, **Scikit-learn** (Random Forest)

Let's dive in! 🛳️


```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier

# Load the dataset
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic_data = pd.read_csv(url)

# Display first few rows of the dataset
titanic_data.head()
```

### Overview  
The Titanic dataset contains information about passengers on the Titanic, including their survival status, class, age, and more. We’ll start by exploring and cleaning the data.


```python
# Check the structure of the dataset
titanic_data.info()

# Check for missing values
titanic_data.isnull().sum()

# Summary statistics of numerical features
titanic_data.describe()
```

### Summary  
- The dataset contains several columns with missing values, including `Age`, `Cabin`, and `Embarked`.  
- We will handle these missing values, remove outliers, and perform feature engineering to create new variables.


```python
# Handling missing Age values using group-based imputation
titanic_data['Age'] = titanic_data.groupby(['Pclass', 'Sex'])['Age'].transform(lambda x: x.fillna(x.median()))

# Fill missing Embarked values with the most common value
titanic_data['Embarked'].fillna(titanic_data['Embarked'].mode()[0], inplace=True)

# Dropping the Cabin column
titanic_data.drop('Cabin', axis=1, inplace=True)

# Check the data again
titanic_data.isnull().sum()
```

### Explanation  
- **Age**: Missing `Age` values are filled based on passenger class (`Pclass`) and gender (`Sex`).  
- **Embarked**: Missing values in `Embarked` are filled with the most common port.  
- **Cabin**: We removed the `Cabin` column due to excessive missing data.




```python
# Handling outliers in the 'Fare' column using IQR method
Q1 = titanic_data['Fare'].quantile(0.25)
Q3 = titanic_data['Fare'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Cap the outliers
titanic_data['Fare'] = np.where(titanic_data['Fare'] > upper_bound, upper_bound, titanic_data['Fare'])

# Visualize the distribution after handling outliers
sns.histplot(titanic_data['Fare'], kde=True)
plt.title('Distribution of Fare after Outlier Handling')
plt.show()
```

### Explanation  
We used the **Interquartile Range (IQR)** method to identify and cap extreme `Fare` values to reduce the influence of outliers.


```python
# Creating FamilySize and IsAlone features
titanic_data['FamilySize'] = titanic_data['SibSp'] + titanic_data['Parch'] + 1
titanic_data['IsAlone'] = np.where(titanic_data['FamilySize'] == 1, 1, 0)

# Extracting Title from Name
titanic_data['Title'] = titanic_data['Name'].apply(lambda x: x.split(',')[1].split('.')[0].strip())

# Simplifying rare titles
rare_titles = ['Lady', 'Countess', 'Capt', 'Col', 'Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona']
titanic_data['Title'] = titanic_data['Title'].replace(rare_titles, 'Rare')
titanic_data['Title'] = titanic_data['Title'].replace(['Mlle', 'Ms'], 'Miss')
titanic_data['Title'] = titanic_data['Title'].replace('Mme', 'Mrs')

# Display the first few rows of the modified dataset
titanic_data[['Name', 'Title', 'FamilySize', 'IsAlone']].head()
```

### Explanation  
- **FamilySize**: The total number of family members on board, including the passenger.  
- **IsAlone**: A binary feature indicating whether the passenger was traveling alone.  
- **Title**: Extracted from the passenger's name to capture social status, which might influence survival.


```python
# Select only numeric columns for correlation matrix
numeric_columns = titanic_data.select_dtypes(include=[np.number])

# Compute the correlation matrix
corr_matrix = numeric_columns.corr()

# Plot the correlation matrix using a heatmap
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Numeric Features')
plt.show()
```

```python
# Survival rate by gender
sns.barplot(x='Sex', y='Survived', data=titanic_data)
plt.title('Survival Rate by Gender')
plt.show()

# Survival rate by passenger class
sns.barplot(x='Pclass', y='Survived', data=titanic_data)
plt.title('Survival Rate by Passenger Class')
plt.show()

# Survival rate by family size
sns.barplot(x='FamilySize', y='Survived', data=titanic_data)
plt.title('Survival Rate by Family Size')
plt.show()
```

### Key Insights  
- **Gender**: Females had a significantly higher survival rate than males.  
- **Class**: Passengers in 1st class had the highest survival rate, while those in 3rd class had the lowest.  
- **FamilySize**: Having a small family (2-4 people) increased the chances of survival, while being alone reduced the likelihood.


```python
# One-hot encode 'Sex' and 'Embarked' (optional if used in the dataset)
titanic_data['Sex'] = titanic_data['Sex'].map({'male': 0, 'female': 1})

# One-hot encode the 'Title' column
titanic_data = pd.get_dummies(titanic_data, columns=['Title'], drop_first=True)

# Select features and target variable
X = titanic_data[['Pclass', 'Age', 'Fare', 'FamilySize', 'IsAlone', 'Sex']]  # You can add encoded Title columns here too
y = titanic_data['Survived']

# Initialize and fit the Random Forest model
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X, y)

# Feature importance plot
feature_importances = pd.Series(rf.feature_importances_, index=X.columns)
feature_importances.sort_values(ascending=False).plot(kind='barh')
plt.title('Feature Importance from Random Forest')
plt.show()
```

### Explanation  
The **Random Forest** model highlights the most important features for predicting survival. We observe that **Sex**, **Pclass**, and **Fare** are the most influential factors.
